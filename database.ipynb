```python
#Use MySql Workbench to create tables
CREATE TABLE `host` (
  `host_id` int(11) NOT NULL,
  `host_name` varchar(200) DEFAULT NULL,
  PRIMARY KEY (`host_id`)
) ENGINE=InnoDB DEFAULT CHARSET=utf8;

CREATE TABLE `house` (
  `id` int(10) unsigned zerofill NOT NULL,
  `name` varchar(200) DEFAULT NULL,
  `latitude` float DEFAULT NULL,
  `longitude` float DEFAULT NULL,
  `room_type` varchar(200) DEFAULT NULL,
  `price` varchar(200) DEFAULT NULL,
  `minimum_nights` int(11) DEFAULT NULL,
  `number_of_reviews` int(11) DEFAULT NULL,
  `last_review` varchar(45) DEFAULT NULL,
  `reviews_per_month` float DEFAULT NULL,
  `calculated_host_listings_count` int(11) DEFAULT NULL,
  `availability_365` int(11) DEFAULT NULL,
  PRIMARY KEY (`id`)
) ENGINE=InnoDB DEFAULT CHARSET=utf8;

CREATE TABLE `locate` (
  `id` int(11) NOT NULL,
  `neighbourhood` varchar(200) DEFAULT NULL,
  PRIMARY KEY (`id`)
) ENGINE=InnoDB DEFAULT CHARSET=utf8;

CREATE TABLE `position` (
  `neighbourhood` varchar(200) NOT NULL,
  `neighbourhood_group` varchar(200) DEFAULT NULL,
  PRIMARY KEY (`neighbourhood`)
) ENGINE=InnoDB DEFAULT CHARSET=utf8;

CREATE TABLE `possess` (
  `id` int(11) NOT NULL,
  `host_id` int(11) DEFAULT NULL,
  PRIMARY KEY (`id`)
) ENGINE=InnoDB DEFAULT CHARSET=utf8;
```

```python
#Delete the "'"s to make it more easier to create sql commands.
#We can also use '' to replace ' sy
for index,row in df.iterrows():
    a=str(row['name']).split('\'')
    df['name'][index]=' '.join(a)
    b=str(row['neighbourhood']).split('\'')
    df['neighbourhood'][index]=' '.join(b)
```
